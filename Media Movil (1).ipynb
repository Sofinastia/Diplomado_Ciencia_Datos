La media movil es el promedio de un valor en un numero de periodos determinados. Así sirven como indicadores tecnicos de tendencia de series de tiempo y a su vez para suavizar estan ante la presencia de tendencia. Se pueden encontrar tres tipos de medeia movil:

- **Media Movil Simple (SMA):** Es una media aritmética de X datos anteriores. Su formula seria \
\
$$SMA=\frac{X_1+X_2+X_3+...+X_n}{n}$$
\
Donde $X_1,...,X_n$ son los valores de la serie temporal y $n$ es el numero de periodos escogido.

- **Media Movil Exponencial (EMA):** Es una media aritmética simple ponderada por X valores anteriores con factores que van decreciendo de forma exponencial. A diferencia de la anterior da más peso a los puntos de datos más recientes, haciendo que reaccione más rápidamente a las variaciones en la serie temporal. Su formula sería\
    \
    $$EMA_t=(1-\alpha)*EMA_{t-1}+\alpha*X_t$$
    \
   Donde $EMA_t$ es la media movil exponencial en el tiempo $t$, $X_t$ es el valor actual en el tiempo $t$ y $\alpha$ es el factor de suavizado que controla la podneración de los valores anteriores en el calculo. 

- **Medial Movil Ponderada(MA):** Es una media multiplicada por diferentes factores, dado que aplican diferente peso a los datos de la serie, así es una variante de la media movil simple. Su formula seria\
    \
    $$MA_t=w_1*X_{t-1}+w_2*X_{t-2}+...+w_n*X_{t-n}$$
    \
    Donde $MA_t$ es la media movil ponderada en el tiempo $t$, $X_{t-1},...,X_{t-n}$ son los valores de la serie temporal en tiempos anteriores y $w_1,...,w_n$ son los pesos asignados a los valores respectivos. Cabe aclarar que la suma de todos los $w$ debe ser igual a $1$.

A continuación se hará un ejemplo de como utilizar le media movil simple y la media movil ponderada

```python
!pip install pandas
!pip install numpy
!pip install matplotlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Crear una serie temporal de ejemplo
np.random.seed(42)
serie_temporal = pd.Series(np.random.randn(100), index=pd.date_range('2023-01-01', periods=100))

# Calcular la media móvil simple (SMA) con una ventana de 5 días
sma = serie_temporal.rolling(window=5).mean()

# Calcular la media móvil ponderada (WMA) con pesos [0.1, 0.2, 0.3, 0.2, 0.1]
weights = np.array([0.1, 0.2, 0.3, 0.2, 0.1])
wma = serie_temporal.rolling(window=5).apply(lambda x: np.dot(x, weights))

# Graficar la serie temporal y las medias móviles
plt.figure(figsize=(10, 6))
plt.plot(serie_temporal, label='Serie Temporal Original', color='blue')
plt.plot(sma, label='SMA (Ventana 5)', color='green')
plt.plot(wma, label='WMA (Ventana 5, Pesos [0.1, 0.2, 0.3, 0.2, 0.1])', color='orange')

plt.title('Media Móvil Simple y Ponderada')
plt.legend()
plt.show()
```

En el ejemplo anterior se creo una serie temporal de 100 valores aleatorios, luego se calculo la media móvil simple (SMA) utilizando la función rolling de pandas con una ventana de 5 días; también se calcula la media móvil ponderada (WMA) utilizando una función lambda y la función rolling de pandas, aplicando los pesos [0.1, 0.2, 0.3, 0.2, 0.1]. Finalmente, se graficó la serie temporal original junto con las medias móviles para visualizar cómo se suavizan las fluctuaciones con estas técnicas.
